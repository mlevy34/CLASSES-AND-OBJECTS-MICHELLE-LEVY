## Short reminders before you begin
A **class** is a blueprint.  
An **object** is one instance created from that blueprint.  
The `__init__` method builds the starting state of each object.  
An **instance attribute** belongs to one object, usually written with `self.attribute_name`.  
A **class attribute** belongs to the class itself and is shared unless shadowed by an instance.  
A **method** is a function inside a class.  
When an object is passed into a method, the method receives a reference to that object.  
If the method **mutates the object's attributes**, the caller sees those changes after the method ends.  
If the method **reassigns the parameter variable** to a different object, that reassignment only changes the local variable inside the method.

### Question C1 — Create a simple class

Create a class named `Book`.

#### Requirements
- The constructor should accept:
  - `title`
  - `author`
  - `pages`
- Store each value in an instance attribute.
- Then create an object named `b1` with your own sample data.
- Print the title and author on one line.
- use `__str__`

**Reminder:** Each object should store its own instance data.

```python

# Write your answer here
class Book:
  def __init__(self, title : str, author : str, pages : int)->None:
    self.title = title
    self.author = author
    self.pages = pages
  def __str__(self)->str:
    return f"Title: {self.title} Author: {self.author}"

b1 = Book("Living Emunah", "Rabbi David Ashear", 300)
print(b1)
```

### Question C4 — One object uses another object

Create two classes: `Player` and `Coach`.

#### `Player`
- constructor should accept:
  - `name`
  - `points`

#### `Coach`
- add a method `praise_player(player_obj)` that returns a string like:
  - `"Great job, Maya! You scored 18 points."`

Create a player object and a coach object, then call the method.

**Reminder:** This question practices passing an object into another object's method and using its data.

```python

# Write your answer here
class Player:
  def __init__(self, name : str, points : int)->None:
    self.name = name
    self.points = points

class Coach:
  def praise_player(self, player_obj : object)->str:
    return f"Great job, {player_obj.name}! You scored {player_obj.points} points."

player = Player("Max", 19)
coach = Coach()
print(coach.praise_player(player))
```
