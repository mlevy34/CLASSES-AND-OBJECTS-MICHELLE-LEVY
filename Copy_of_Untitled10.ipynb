```python
class Flower:
  def __init__(self, name, petals, color, height):
    self.name = name
    self.petals = petals
    self.color = color
    self.height = height
  def describe(self):
    return f"{self.name} is {self.color}, has {self.petals} petals, and is {self.height} cm tall"
  def grow(self):
    self.height += 2
flowers_lst = []

flower1 = Flower("Lily", 5, "Pink", 14)
flowers_lst.append(flower1)

flower2 = Flower("Rose", 10, "Red", 30)
flowers_lst.append(flower2)

flower3 = Flower("Daisy", 16, "Yellow", 12)
flowers_lst.append(flower3)

flower4 = Flower("Tulip", 8, "White", 24)
flowers_lst.append(flower4)

for flower in flowers_lst:
  print(flower.describe())
print()
for flower in flowers_lst:
  flower.grow()

for flower in flowers_lst:
  print(flower.describe())

```
