## Big Reminders
A **class** is a blueprint for making objects.  
An **object** is one instance of that class.  
The constructor `__init__` gives each object its starting values.  
Instance attributes usually belong to one object.  
Class attributes belong to the class and are shared unless shadowed.  
Methods let objects do work.  
One object can receive another object as a parameter and use that object's data.  
If a method mutates an object, that mutation is still there when the method returns.  
If a method reassigns the parameter variable to a different object, the original caller's object does not get replaced.

## Task 9 — A `BankAccount` class

Create a class named `BankAccount`.

### Requirements
- constructor parameters:
  - `owner`
  - `balance`
- method `deposit(amount)`
  - add to the balance
- method `withdraw(amount)`
  - if enough money exists, subtract and return `True`
  - otherwise do not change the balance and return `False`
- method `display()`
  - return a string like:
    - `"Ari has $120."`

Create one account and test all methods.

### Reminder
This is classic object-oriented design: the object stores state and methods manage that state.

````python

# Write your answer here
class BankAccount:
  def __init__(self, owner: str, balance: float) -> None:
    """Initialize bank account with owner name and starting balance."""
    self.owner = owner
    self.balance = balance

  def deposit(self, amount: int) -> None:
    """Add money to the account balance."""
    # increase balance by deposit amount
    self.balance += amount

  def withdraw(self, amount: int) -> bool:
    """Withdraw money if sufficient balance exists."""
    # check if there are enough funds
    if self.balance >= amount:
      # subtract amount from balance
      self.balance -= amount
      return True
    else:
      # withdrawal fails if insufficient funds
      return False

  def display(self) -> str:
    """Return formatted account information."""
    # return a string showing owner and balance
    return f"{self.owner} has ${self.balance}"

# TESTS
b1 = BankAccount("Michelle", 500)
b1.deposit(100)
print(b1.withdraw(700))
b1.display()
````

## Task 10 — A `Classroom` and `Student`

Create two classes: `Student` and `Classroom`.

### `Student`
- constructor parameters:
  - `name`
  - `grade`

### `Classroom`
- constructor should create an empty list named `students`
- method `add_student(student_obj)`
  - add the object to the list
- method `highest_grade()`
  - return the highest grade among the students
  - if there are no students, return `None`

### Test
- create at least 3 students
- add them to a classroom
- print the highest grade

### Reminder
This is good practice with objects containing collections of other objects.

````python

# Write your answer here
class Student:
  def __init__(self, name: str, grade: int) -> None:
    """Initialize a student with a name and grade."""
    self.name = name
    self.grade = grade


class Classroom:
  def __init__(self) -> None:
    """Initialize an empty classroom list."""
    self.students = []

  def add_student(self, student_obj: object) -> None:
    """Add a student object to the classroom."""
    # append student object into students list
    self.students.append(student_obj)

  def highest_grade(self) -> int:
    """Return the highest grade in the classroom."""
    # if no students exist, return None
    if len(self.students) == 0:
      return None
    else:
      grades = []  # store all student grades

      # loop through each student object
      for student in self.students:
        # extract grade and add to list
        grades.append(student.grade)

      # return highest grade found
      return max(grades)
# TESTS
s1 = Student("Michelle", 99)
s2 = Student("Sarah", 97)
s3 = Student("Leah", 91)

classroom = Classroom()
classroom.add_student(s1)
classroom.add_student(s2)
classroom.add_student(s3)

classroom.highest_grade()


````

## Task 11 — Design + object interaction challenge

Create two classes: `Player` and `InventoryManager`.

### `Player`
- constructor parameters:
  - `name`
  - `items`  (this will be a list of strings)

### `InventoryManager`
Add two methods:

1. `add_item(player_obj, item)`
   - append the item to the player's `items` list

2. `trade_first_item(player_a, player_b)`
   - if **both** players have at least one item:
     - swap their first items
     - return `True`
   - otherwise return `False`

### Test
Create two players with different item lists.
Demonstrate both methods.

### Why this is a thinking question
You must manage state stored inside objects, including a mutable list attribute.
Be careful to mutate the correct object data.

````python

# Write your answer here
class Player:
  def __init__(self, name, items) -> None:
    """Initialize a player with a name and a list of items."""
    self.name = name
    self.items = items


class InventoryManager:
  def add_item(self, player_obj: object, item: str) -> None:
    """Add an item to a player's inventory."""
    # add item to the player's list
    player_obj.items.append(item)

  def trade_first_item(self, player_a: object, player_b: object) -> bool:
    """Swap the first item between two players if both have items."""
    # ensure both players have at least one item
    if len(player_a.items) > 0 and len(player_b.items) > 0:
      # store first items temporarily
      first1 = player_a.items[0]
      first2 = player_b.items[0]

      # swap first items
      player_a.items[0] = first2
      player_b.items[0] = first1

      return True
    else:
      # trade not possible if either player has no items
      return False

# TESTS
player1 = Player("Michelle", ["sword", "sheild", "dress"])
player2 = Player("Sarah", ["knife", "torch", "gun"])
print(f"Original inventory: player 1:{player1.items} player 2: {player2.items}")
inventory = InventoryManager()
inventory.add_item(player1, "needle")
inventory.add_item(player2, "saw")
print(f"Inventory after items added: player 1: {player1.items} player 2:{player2.items}")
inventory.trade_first_item(player1, player2)
print(f"Inventory after items swapped: player 1: {player1.items} player 2:{player2.items}")
````

## Task 12 — LeetCode-style object design challenge

Create a class named `ScoreTracker`.

### Requirements
- constructor creates an empty dictionary to store player scores
- method `add_score(player_name, points)`
  - if the player is new, create the entry
  - otherwise add to the existing score
- method `top_player()`
  - return the name of the player with the highest score
  - if there are no players, return `None`
- method `top_score()`
  - return the highest score
  - if there are no players, return `None`

### Test
Add several scores for several players and show that repeated calls accumulate correctly.

### Reflection
In a comment, explain why this is a good example of a class:
- what state is being stored?
- what behaviors are being grouped with that state?

````python
# Write your answer here
class ScoreTracker:
  def __init__(self)->None:
    """Initialize storage for player scores."""
    self.scores = {}  # dictionary mapping player names to their total points

  def add_score(self, player_name : str, points : int)->None:
    """Add points to a player, creating them if needed."""
    # if player does not exist yet, create entry with initial points
    if player_name not in self.scores:
      self.scores[player_name] = points
    else:
      # if player already exists, increase their score
      self.scores[player_name] += points

  def top_player(self)->str:
    """Return the player with the highest score."""
    # if no players exist, return None
    if self.scores == {}:
      return None
    else:
      highest = 0  # keep track of the highest score found so far
      best = ""    # keep track of the player with the highest score

      # loop through all players and their scores
      for key, value in self.scores.items():
        # update if a higher score is found
        if value > highest:
          highest = value
          best = key
      return best  # return player with highest score

  def top_score(self)->int:
    """Return the highest score."""
    # if no players exist, return None
    if self.scores == {}:
      return None
    else:
      highest = 0  # keep track of the highest score

      # loop through all scores
      for value in self.scores.values():
        # update if current score is higher
        if value > highest:
          highest = value
      return highest  # return highest score

# TESTS
scoretracker = ScoreTracker()
scoretracker.add_score("michelle", 96)
scoretracker.add_score("sarah", 98)
scoretracker.add_score("raizy", 95)
scoretracker.add_score("rebecca", 93)
scoretracker.add_score("randy", 83)
print(scoretracker.top_player())
print(scoretracker.top_score())

````
