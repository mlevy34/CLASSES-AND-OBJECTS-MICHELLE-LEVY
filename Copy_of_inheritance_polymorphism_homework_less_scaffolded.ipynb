## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
# Write your code for Problem 1 here.
class Animal:
  def __init__(self, name : str, age : int = 5)->None:
    self.name = name
    self.age = age
  def speak(self):
    return "wooo"
  def move(self):
    return "walked 5 steps"
  def describe(self):
    return "this animal is a living being"

class Dog(Animal):
  def __init__(self):
    super().__init__("dog", 10)
  def speak(self):
    return "ruf ruf"
  def move(self):
    return "The dog jumped"

class Cat(Animal):
  def __init__(self):
    super().__init__("cat", 8)
  def speak(self):
    return "meow meow"
  def move(self):
    return "The cat jumped"

class Rabbit(Animal):
  def __init__(self):
    super().__init__("rabbit", 2)
  def speak(self):
    return "ribbet ribbet"
  def move(self):
    return "The rabbit jumped"

dog = Dog()
cat = Cat()
rabbit = Rabbit()


print(f"Name: {dog.name} Age: {dog.age}")
print(f"Name: {cat.name} Age: {cat.age}")
print(f"Name: {rabbit.name} Age: {rabbit.age}")
print()
print(dog.speak())
print(cat.speak())
print(rabbit.speak())
print()
print(dog.move())
print(cat.move())
print(rabbit.move())
print()
print(dog.describe())
print(cat.describe())
print(rabbit.describe())


```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
# Write your code for Problem 2 here.
class Vehicle:
  def __init__(self, brand : str, speed : int = 5)->None:
    self.brand = brand
    self.speed = speed
  def move(self):
    return "transportation is great!"
  def sound_horn(self):
    return "honk"
  def describe(self):
    return "this vehicle moves"

class Car(Vehicle):
  def __init__(self):
    super().__init__("toyota", 10)
  def move(self):
    return "this is a car"
  def sound_horn(self):
    return "Honk car coming!"

class Truck(Vehicle):
  def __init__(self):
    super().__init__("nissan", 5)
  def move(self):
    return "this is a truck"
  def sound_horn(self):
    return "Honk truck coming!"

class Motorcycle(Vehicle):
  def __init__(self):
    super().__init__("volvo", 20)
  def move(self):
    return "this is a motorcycle"
  def sound_horn(self):
    return "Honk motorcycle coming!"

car = Car()
truck = Truck()
motorcycle = Motorcycle()

print(f"Brand: {car.brand} Speed: {car.speed}")
print(f"Brand: {truck.brand} Speed: {truck.speed}")
print(f"Brand: {motorcycle.brand} Speed: {motorcycle.speed}")
print()
print(car.move())
print(truck.move())
print(motorcycle.move())
print()
print(car.sound_horn())
print(truck.sound_horn())
print(motorcycle.sound_horn())
print()
print(car.describe())
print(truck.describe())
print(motorcycle.describe())
```


## Problem 3: Demonstrate Polymorphism with Animals

Using your `Dog`, `Cat`, and `Rabbit` classes:

1. Create a list that stores multiple animal objects together.
2. Use a loop to go through the list.
3. For each object, call the **same methods** and show that the results differ depending on the object.

### Required methods to call in the loop
- `speak()`
- `move()`
- `describe()`

## Goal
Your output should clearly demonstrate that the program can treat all of the child objects as `Animal` objects, while each one still behaves in its own way.


```python
# Write your code for Problem 3 here.
animals = [Dog(), Cat(), Rabbit()]
for animal in animals:
  print(f"Name: {animal.name} Age: {animal.age}")
  print(animal.speak())
  print(animal.move())
  print(animal.describe())
  print()

```
